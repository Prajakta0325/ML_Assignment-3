```python
data = pd.read_csv('mobile_price_data.csv')
data.head()
```

- drop unnecessary column

```python
data = data.drop(['mobile_color', 'dual_sim', 'bluetooth', 'mob_width',
       'mob_height', 'mob_depth', 'mob_weight'],axis=1)
data.head()
```

```python
data.info()
```

### Observation:
- Mobile Price is dependent on other variable so it is dependent/Target variable.

```python
data.columns
```

```python
data['brand']=data.mobile_name.str.extract(r"(\w+)")
```

```python
data['price(Rs.)']=data.mobile_price.str.extract(r"(\d+)")+data.mobile_price.str.extract(r",(\d+)")

data['display(cm)']=data.disp_size.str.extract(r"([0-9\.]+)")

data['res']=data.resolution.str.extract(r"(720|1080|540)")

data['cores']=data.num_cores.map({'Octa Core':8, 'Quad Core':4, 'Single Core':1})

data['clock_speed(GHz)']=data.mp_speed.str.extract(r"([0-9\.])")

data['storage']=data.int_memory.str.extract(r"(\d+)")

data['ram']=data.ram.str.extract(r"(\d+)")

data['primary_cam(MP)']=data.p_cam.str.extract(r"(\d+)")

data['back_camera_count']=data.p_cam.str.count(r"(\d+)")

data['front_cam(MP)']=data.f_cam.str.extract(r"(\d+)")

data['battery(Mah)']=data.battery_power.str.extract(r"(\d+)")
```

```python
data = data.drop([ 'mobile_name','mobile_price', 'disp_size', 'resolution', 'os',
       'num_cores', 'mp_speed', 'int_memory', 'ram', 'p_cam', 'f_cam',
       'network', 'battery_power'],axis=1)
data.head()
```

```python
data.info()
```

```python
data['brand'].unique()
```

```python
# apply one-hot encoding on text data
```

```python
data_ohe = pd.get_dummies(data['brand'])
data_ohe
```

```python
data = pd.concat([data,data_ohe], axis=1)
data.head()
```

```python
data= data.drop("brand",axis=1)
data.head()
```

```python
### Step 3: Separate X and Y
```

```python
x= data.drop('price(Rs.)',axis=1)
y= data['price(Rs.)']
```

```python
### Step 4: Split data into ttrain and test set
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
### Step 5: Apply Linear Regression algorithm
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr
```

```python
lr.fit(x_train,y_train)
```

```python
### Step 6: Prediction on test data
```

```python
y_pred = lr.predict(x_test)
y_pred
```

```python
### Step 8: Evaluation
```

```python
from sklearn.metrics import r2_score
accuracy = r2_score(y_test,y_pred)
accuracy
```

- dataset contain 70% accuracy

```python
# LR equation = y = mx + c
```

```python
m = lr.coef_
m
```

```python
c = lr.intercept_
c
```
